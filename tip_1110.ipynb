```python
import pandas as pd

# datetime 형태로 변환
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])

# 시(hour)만 추출하는 새 컬럼 생성
data['hour'] = data['tpep_pickup_datetime'].dt.hour
```

```python
hourly_passengers = data.groupby('hour')['passenger_count'].sum().reset_index()
hourly_passengers
```

```python
#승객이 가장 많은 시간 추출
```

```python
hourly_passengers.loc[hourly_passengers['passenger_count'].idxmax()]
```
