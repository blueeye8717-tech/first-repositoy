### ğŸ’¡ í•µì‹¬ ê°œë…:

**í—ˆìœ„ ìƒê´€**ì€ ë‘ ë³€ìˆ˜ ì‚¬ì´ì— ì§ì ‘ì ì¸ ì¸ê³¼ê´€ê³„ê°€ ì—†ì§€ë§Œ, 'ê¸°ì˜¨'ê³¼ ê°™ì€ ì œ3ì˜ ë³€ìˆ˜(ì¤‘ì²©ìš”ì¸)ê°€ ë‘ ë³€ìˆ˜ ëª¨ë‘ì— ì˜í–¥ì„ ì£¼ì–´ ë§ˆì¹˜ ê´€ê³„ê°€ ìˆëŠ” ê²ƒì²˜ëŸ¼ ë³´ì´ëŠ” í˜„ìƒì…ë‹ˆë‹¤. **ë‹¤ì¤‘íšŒê·€ë¶„ì„**ì— ì¤‘ì²©ìš”ì¸ì„ í•¨ê»˜ íˆ¬ì…í•˜ë©´, ë‹¤ë¥¸ ë³€ìˆ˜ë“¤ì˜ ì˜í–¥ì„ ê³ ì •í•œ ìƒíƒœì—ì„œì˜ 'ìˆœìˆ˜í•œ' ê´€ê³„ë¥¼ íŒŒì•…í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.


### ğŸ“Œ ìˆ˜í–‰ ê³¼ì œ:

1.  ê°€ìƒì˜ ë°ì´í„°ë¥¼ ìƒì„±í•©ë‹ˆë‹¤: 'ê¸°ì˜¨'ì´ ë†’ì•„ì§ˆìˆ˜ë¡ 'ì•„ì´ìŠ¤í¬ë¦¼ íŒë§¤ëŸ‰'ê³¼ 'í•´ìˆ˜ìš•ê° ìˆ˜'ê°€ ëª¨ë‘ ì¦ê°€í•˜ëŠ” ìƒí™©ì„ ì‹œë®¬ë ˆì´ì…˜í•©ë‹ˆë‹¤.
2.  **ë‹¨ìˆœíšŒê·€ë¶„ì„**: 'ì•„ì´ìŠ¤í¬ë¦¼ íŒë§¤ëŸ‰'ë§Œìœ¼ë¡œ 'í•´ìˆ˜ìš•ê° ìˆ˜'ë¥¼ ì˜ˆì¸¡í•˜ëŠ” ëª¨ë¸ì„ ë§Œë“­ë‹ˆë‹¤. ê³„ìˆ˜ê°€ ìœ ì˜ë¯¸í•˜ê²Œ ë‚˜ì˜¤ëŠ”ì§€ í™•ì¸í•©ë‹ˆë‹¤.
3.  **ë‹¤ì¤‘íšŒê·€ë¶„ì„**: 'ì•„ì´ìŠ¤í¬ë¦¼ íŒë§¤ëŸ‰'ê³¼ 'ê¸°ì˜¨'ì„ ëª¨ë‘ ì‚¬ìš©í•˜ì—¬ 'í•´ìˆ˜ìš•ê° ìˆ˜'ë¥¼ ì˜ˆì¸¡í•˜ëŠ” ëª¨ë¸ì„ ë§Œë“­ë‹ˆë‹¤.
4.  ë‘ ëª¨ë¸ì˜ ê²°ê³¼ë¥¼ ë¹„êµí•˜ê³ , 'ì•„ì´ìŠ¤í¬ë¦¼ íŒë§¤ëŸ‰'ì˜ ê³„ìˆ˜(coef)ì™€ p-valueê°€ ì–´ë–»ê²Œ ë³€í–ˆëŠ”ì§€ í•´ì„í•´ ë³´ì„¸ìš”.


```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# 1. ê°€ìƒ ë°ì´í„° ìƒì„± (ì´ ì½”ë“œëŠ” ìˆ˜ì •í•˜ì§€ ë§ˆì„¸ìš”)
np.random.seed(42)
n_samples = 100
temperature = np.random.uniform(15, 35, n_samples)
ice_cream_sales = 2 * temperature + np.random.normal(0, 5, n_samples)
beach_visitors = 10 * temperature + np.random.normal(0, 20, n_samples)
df1 = pd.DataFrame({'ê¸°ì˜¨': temperature, 'ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰': ice_cream_sales, 'í•´ìˆ˜ìš•ê°ìˆ˜': beach_visitors})
print(df1.corr())

# 2. ë‹¨ìˆœíšŒê·€ë¶„ì„: ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰ -> í•´ìˆ˜ìš•ê°ìˆ˜
print("\n--- ëª¨ë¸ 1: ë‹¨ìˆœíšŒê·€ë¶„ì„ (ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰ -> í•´ìˆ˜ìš•ê°ìˆ˜) ---")
y = df1['í•´ìˆ˜ìš•ê°ìˆ˜']

# ë…ë¦½ë³€ìˆ˜ë¡œ 'ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰'ë§Œ ì„ íƒ
X_simple = df1[['ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰']]

# ìƒìˆ˜í•­(ì ˆí¸) ì¶”ê°€
X_simple_const = sm.add_constant(X_simple)

# OLS ëª¨ë¸ í•™ìŠµ
model_simple = sm.OLS(y, X_simple_const).fit()
print(model_simple.summary())

# 3. ë‹¤ì¤‘íšŒê·€ë¶„ì„: (ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰, ê¸°ì˜¨) -> í•´ìˆ˜ìš•ê°ìˆ˜
print("\n--- ëª¨ë¸ 2: ë‹¤ì¤‘íšŒê·€ë¶„ì„ (ì¤‘ì²©ìš”ì¸ 'ê¸°ì˜¨' í†µì œ) ---")

# ë…ë¦½ë³€ìˆ˜ë¡œ 'ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰'ê³¼ 'ê¸°ì˜¨' ëª¨ë‘ ì„ íƒ
X_multi = df1[['ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰', 'ê¸°ì˜¨']]

# ìƒìˆ˜í•­ ì¶”ê°€
X_multi_const = sm.add_constant(X_multi)

# OLS ëª¨ë¸ í•™ìŠµ
model_multi = sm.OLS(y, X_multi_const).fit()
print(model_multi.summary())

# 4. ê²°ê³¼ ë¹„êµ: coefì™€ p-valueë§Œ ê¹”ë”í•˜ê²Œ ë½‘ê¸°
print("\n--- coef / p-value ë¹„êµ (ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰) ---")
print("ëª¨ë¸1 coef:", model_simple.params['ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰'])
print("ëª¨ë¸1 p-value:", model_simple.pvalues['ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰'])

print("ëª¨ë¸2 coef:", model_multi.params['ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰'])
print("ëª¨ë¸2 p-value:", model_multi.pvalues['ì•„ì´ìŠ¤í¬ë¦¼íŒë§¤ëŸ‰'])
```

### ğŸ¤” ìƒê°í•´ ë³¼ ë¬¸ì œ:

## ë¬¸ì œ 2 (ë‚œì´ë„: í•˜): í‘œë³¸ í¬ê¸°ì˜ í•¨ì •ê³¼ íš¨ê³¼ í¬ê¸°ì˜ ì¤‘ìš”ì„±

> **ğŸ¯ ëª©í‘œ:** í‘œë³¸ í¬ê¸°(n)ê°€ ì»¤ì§€ë©´ ì•„ì£¼ ë¯¸ë¯¸í•œ ì°¨ì´ë„ í†µê³„ì ìœ¼ë¡œ ìœ ì˜(p<0.05)í•˜ê²Œ ë˜ëŠ” í˜„ìƒì„ í™•ì¸í•˜ê³ , ì´ ë•Œë¬¸ì— **íš¨ê³¼ í¬ê¸°(Effect Size)**ë¥¼ í•¨ê»˜ í™•ì¸í•´ì•¼ í•˜ëŠ” ì´ìœ ë¥¼ ì´í•´í•©ë‹ˆë‹¤.


### ğŸ’¡ í•µì‹¬ ê°œë…:

ê°€ì„¤ê²€ì •ì—ì„œ p-valueëŠ” 'ê·€ë¬´ê°€ì„¤ì´ ë§ë‹¤ê³  ê°€ì •í•  ë•Œ, í˜„ì¬ ë°ì´í„°ì™€ ê°™ê±°ë‚˜ ë” ê·¹ë‹¨ì ì¸ ê²°ê³¼ê°€ ë‚˜ì˜¬ í™•ë¥ 'ì…ë‹ˆë‹¤. í‘œë³¸ì´ ì»¤ì§ˆìˆ˜ë¡ ì•„ì£¼ ì‘ì€ ì°¨ì´ë„ ìš°ì—°ìœ¼ë¡œ ë³´ê¸° ì–´ë ¤ì›Œì§€ë¯€ë¡œ p-valueëŠ” ì‘ì•„ì§‘ë‹ˆë‹¤. **íš¨ê³¼ í¬ê¸°**ëŠ” ì´ ì°¨ì´ê°€ 'ì‹¤ì§ˆì ìœ¼ë¡œ ì–¼ë§ˆë‚˜ ì˜ë¯¸ ìˆëŠ” í¬ê¸°ì¸ê°€'ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ì§€í‘œë¡œ, p-valueì˜ í•œê³„ë¥¼ ë³´ì™„í•´ ì¤ë‹ˆë‹¤.


### ğŸ“Œ ìˆ˜í–‰ ê³¼ì œ:

1.  í‰ê· ì´ ì•„ì£¼ ì•½ê°„ ë‹¤ë¥¸(ì˜ˆ: 170cm vs 170.1cm) ë‘ ì§‘ë‹¨ì„ ê°€ì •í•©ë‹ˆë‹¤.
2.  **ì‘ì€ í‘œë³¸(n=30)**ì„ ê° ì§‘ë‹¨ì—ì„œ ì¶”ì¶œí•˜ì—¬ t-ê²€ì •ì„ ìˆ˜í–‰í•˜ê³  p-valueë¥¼ í™•ì¸í•©ë‹ˆë‹¤.
3.  **í° í‘œë³¸(n=10,000)**ì„ ê° ì§‘ë‹¨ì—ì„œ ì¶”ì¶œí•˜ì—¬ t-ê²€ì •ì„ ìˆ˜í–‰í•˜ê³  p-valueë¥¼ í™•ì¸í•©ë‹ˆë‹¤.
4.  ë‘ ê²½ìš° ëª¨ë‘ì— ëŒ€í•´ **íš¨ê³¼ í¬ê¸°(Cohen's d)**ë¥¼ ì§ì ‘ ê³„ì‚°í•˜ê³ , p-valueì™€ íš¨ê³¼ í¬ê¸°ê°€ ì–´ë–»ê²Œ ë‹¤ë¥¸ ë©”ì‹œì§€ë¥¼ ì£¼ëŠ”ì§€ ë¹„êµ ë¶„ì„í•˜ì„¸ìš”.


```python
import numpy as np
from scipy import stats

# íš¨ê³¼ í¬ê¸°(Cohen's d) ê³„ì‚° í•¨ìˆ˜ (ì´ ì½”ë“œëŠ” ìˆ˜ì •í•˜ì§€ ë§ˆì„¸ìš”)
def cohen_d(x, y):
    nx, ny = len(x), len(y)
    dof = nx + ny - 2
    return (np.mean(x) - np.mean(y)) / np.sqrt(((nx-1)*np.std(x, ddof=1)**2 + (ny-1)*np.std(y, ddof=1)**2) / dof)

np.random.seed(123)
mu1, mu2, sigma = 170, 170.1, 5

# 2. ì‘ì€ í‘œë³¸(n=30)ìœ¼ë¡œ ê²€ì •
sample1_small = np.random.normal(mu1, sigma, 30)
sample2_small = np.random.normal(mu2, sigma, 30)

# t-ê²€ì • ìˆ˜í–‰
t_stat_small, p_value_small = stats.ttest_ind(sample1_small, sample2_small)
effect_size_small = cohen_d(sample1_small, sample2_small)

print(f"--- ì‘ì€ í‘œë³¸ (n=30) ê²°ê³¼ ---")
print(f"t-stat: {t_stat_small:.4f}")
print(f"P-value: {p_value_small:.4f}")
print(f"Effect Size (Cohen's d): {effect_size_small:.4f}\n")

# 3. í° í‘œë³¸(n=10000)ìœ¼ë¡œ ê²€ì •
sample1_large = np.random.normal(mu1, sigma, 10000)
sample2_large = np.random.normal(mu2, sigma, 10000)

# t-ê²€ì • ìˆ˜í–‰
t_stat_large, p_value_large = stats.ttest_ind(sample1_large, sample2_large)
effect_size_large = cohen_d(sample1_large, sample2_large)

print(f"--- í° í‘œë³¸ (n=10000) ê²°ê³¼ ---")
print(f"t-stat: {t_stat_large:.4f}")
print(f"P-value: {p_value_large:.4f}")
print(f"Effect Size (Cohen's d): {effect_size_large:.4f}\n")
```

## ë¬¸ì œ 3 (ë‚œì´ë„: ì¤‘): p-í•´í‚¹ ì‹œë®¬ë ˆì´ì…˜

> **ğŸ¯ ëª©í‘œ:** 'ë°ì´í„°ë¥¼ ë³´ë©´ì„œ ë¶„ì„ì„ ê³„ì†í• ì§€ ê²°ì •í•˜ëŠ” í–‰ìœ„'ê°€ ì–´ë–»ê²Œ ì˜ëª»ëœ ê²°ë¡ (ì œ1ì¢… ì˜¤ë¥˜)ì˜ ê°€ëŠ¥ì„±ì„ ë†’ì´ëŠ”ì§€, ì¦‰ **p-í•´í‚¹**ì˜ ìœ„í—˜ì„±ì„ ì‹œë®¬ë ˆì´ì…˜ì„ í†µí•´ ì§ì ‘ ì¦ëª…í•©ë‹ˆë‹¤.


### ğŸ’¡ í•µì‹¬ ê°œë…:

**p-í•´í‚¹**ì€ ì—°êµ¬ìê°€ ì˜ë„í–ˆë“  ì•ˆ í–ˆë“  p-valueë¥¼ ë‚®ì¶”ê¸° ìœ„í•´ ë¶„ì„ ê³¼ì •ì„ ì¡°ì‘í•˜ëŠ” í–‰ìœ„ë¥¼ ë§í•©ë‹ˆë‹¤. ëŒ€í‘œì ì¸ ì˜ˆë¡œ, p-valueê°€ 0.05ë¥¼ ë„˜ìœ¼ë©´ í‘œë³¸ì„ ë” ëª¨ì•„ì„œ ë‹¤ì‹œ ê²€ì •í•˜ëŠ” í–‰ìœ„ê°€ ìˆìŠµë‹ˆë‹¤. ì›ë˜ ì œ1ì¢… ì˜¤ë¥˜(ì‹¤ì œ ì°¨ì´ê°€ ì—†ëŠ”ë°, ì°¨ì´ê°€ ìˆë‹¤ê³  ê²°ë¡  ë‚´ë¦´ ì˜¤ë¥˜) í™•ë¥ ì€ ìœ ì˜ìˆ˜ì¤€ Î±(ë³´í†µ 0.05)ë¡œ í†µì œë˜ì–´ì•¼ í•˜ì§€ë§Œ, p-í•´í‚¹ì€ ì´ í™•ë¥ ì„ Î±ë³´ë‹¤ í›¨ì”¬ ë†’ê²Œ ë§Œë“­ë‹ˆë‹¤.


### ğŸ“Œ ìˆ˜í–‰ ê³¼ì œ:

1.  **ì•„ë¬´ ì°¨ì´ê°€ ì—†ëŠ”** ë‘ ì§‘ë‹¨ A, Bë¥¼ ê°€ì •í•©ë‹ˆë‹¤ (ê·€ë¬´ê°€ì„¤ì´ ì°¸ì¸ ìƒí™©).
2.  'p-í•´ì»¤'ë¥¼ ì‹œë®¬ë ˆì´ì…˜í•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“­ë‹ˆë‹¤. ì´ í•¨ìˆ˜ëŠ” ë‹¤ìŒ ë¡œì§ì„ ë”°ë¦…ë‹ˆë‹¤:
    - ë¨¼ì € ê° ì§‘ë‹¨ì—ì„œ 20ëª…ì”© í‘œë³¸ì„ ì¶”ì¶œí•˜ì—¬ t-ê²€ì •ì„ í•©ë‹ˆë‹¤.
    - p < 0.05ì´ë©´ 'ìœ ì˜ë¯¸í•œ ê²°ê³¼ ë°œê²¬!'ì„ ì™¸ì¹˜ê³  ë©ˆì¶¥ë‹ˆë‹¤.
    - p >= 0.05ì´ë©´, 10ëª…ì”© í‘œë³¸ì„ ì¶”ê°€í•˜ì—¬ ë‹¤ì‹œ t-ê²€ì •ì„ í•©ë‹ˆë‹¤. ì´ ê³¼ì •ì„ ìµœëŒ€ 5ë²ˆê¹Œì§€ ë°˜ë³µí•©ë‹ˆë‹¤.
3.  ì´ ì‹œë®¬ë ˆì´ì…˜ì„ 1,000ë²ˆ ë°˜ë³µí•˜ì—¬, 'p-í•´ì»¤'ê°€ ê²°êµ­ 'ìœ ì˜ë¯¸í•œ ê²°ê³¼'ë¥¼ ë°œê²¬í•  í™•ë¥ ì´ ëª‡ %ì¸ì§€ ê³„ì‚°í•©ë‹ˆë‹¤.
4.  ê³„ì‚°ëœ í™•ë¥ ì´ ì™œ 5%ë³´ë‹¤ í›¨ì”¬ ë†’ì€ì§€ ì„¤ëª…í•˜ì„¸ìš”.


```python
import numpy as np
from scipy import stats

# p-í•´í‚¹ ì‹œë®¬ë ˆì´ì…˜ í•¨ìˆ˜ (ì´ ì½”ë“œëŠ” ìˆ˜ì •í•˜ì§€ ë§ˆì„¸ìš”)
def p_hacking_simulation():
    mu, sigma = 0, 1
    sample_a, sample_b = list(np.random.normal(mu, sigma, 20)), list(np.random.normal(mu, sigma, 20))
    for i in range(5):
        t_stat, p_val = stats.ttest_ind(sample_a, sample_b)
        if p_val < 0.05:
            return True # p-í•´í‚¹ ì„±ê³µ!
        if i < 4:
            sample_a.extend(np.random.normal(mu, sigma, 10))
            sample_b.extend(np.random.normal(mu, sigma, 10))
    return False # 5ë²ˆ ì‹œë„ ëª¨ë‘ ì‹¤íŒ¨

# 3. ì‹œë®¬ë ˆì´ì…˜ 1000ë²ˆ ì‹¤í–‰
n_simulations = 1000
p_hacking_success_count = 0

# forë¬¸ì„ ì‚¬ìš©í•˜ì—¬ ì‹œë®¬ë ˆì´ì…˜ì„ 1000ë²ˆ ë°˜ë³µí•˜ì„¸ìš”.
for _ in range(n_simulations):
    # p_hacking_simulation() í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•˜ê³ , ê²°ê³¼ê°€ Trueì´ë©´ (p-í•´í‚¹ì— ì„±ê³µí•˜ë©´)
    # p_hacking_success_countë¥¼ 1 ì¦ê°€ì‹œí‚¤ì„¸ìš”.
    if p_hacking_simulation():
        p_hacking_success_count += 1

type_1_error_rate = p_hacking_success_count / n_simulations
print(f"ê´€ì°°ëœ ì œ1ì¢… ì˜¤ë¥˜ìœ¨: {type_1_error_rate * 100:.2f}%")
```

ì™œ 5%ë³´ë‹¤ ë†’ê²Œ ë‚˜ì™”ëŠ”ê°€?

ìœ ì˜ìˆ˜ì¤€ 5%ëŠ” ê²€ì •ì„ í•œ ë²ˆë§Œ í–ˆì„ ë•Œì˜ ì˜¤ë¥˜ í™•ë¥ ì´ì§€ë§Œ, p-í•´í‚¹ì€ ë°ì´í„°ë¥¼ ë³´ë©´ì„œ ì—¬ëŸ¬ ë²ˆ ê²€ì •ì„ ë°˜ë³µí•˜ê³  p<0.05ê°€ ë‚˜ì˜¬ ë•Œê¹Œì§€ ë©ˆì¶”ê¸° ë•Œë¬¸ì— ìš°ì—°íˆ ìœ ì˜í•œ ê²°ê³¼ê°€ ë‚˜ì˜¬ í™•ë¥ ì´ ëˆ„ì ë˜ì–´ ì‹¤ì œ ì œ1ì¢… ì˜¤ë¥˜ìœ¨ì´ 5%ë³´ë‹¤ í›¨ì”¬ ë†’ì•„ì§„ë‹¤.

p-í•´í‚¹ì„ ë§‰ê¸° ìœ„í•œ í˜„ì‹¤ì ì¸ ë°©ë²• ì¤‘ í•˜ë‚˜ëŠ” **'ì‚¬ì „ ë“±ë¡(Preregistration)'** ì…ë‹ˆë‹¤. ì—°êµ¬ë¥¼ ì‹œì‘í•˜ê¸° ì „ì— 'í‘œë³¸ í¬ê¸°ëŠ” ëª‡ìœ¼ë¡œ í•  ê²ƒì´ë©°, ì–´ë–¤ ë¶„ì„ì„ í•  ê²ƒì´ë‹¤'ë¼ê³  ë¯¸ë¦¬ ê³µê°œì ìœ¼ë¡œ ì•½ì†í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤. ì´ ë°©ë²•ì´ ìœ„ ì‹œë®¬ë ˆì´ì…˜ì—ì„œ ë³´ì—¬ì¤€ p-í•´í‚¹ ë¬¸ì œë¥¼ ì–´ë–»ê²Œ ë°©ì§€í•  ìˆ˜ ìˆì„ì§€ ì„¤ëª…í•´ ë³´ì„¸ìš”.

## ë¬¸ì œ 4 (ë‚œì´ë„: ì¤‘): ê²½í–¥ ì ìˆ˜ ì§ì§“ê¸°(PSM)ë¡œ ì¸ê³¼ íš¨ê³¼ ì¶”ì •í•˜ê¸°

> **ğŸ¯ ëª©í‘œ:** ë¬´ì‘ìœ„ í†µì œ ì‹¤í—˜(RCT)ì´ ë¶ˆê°€ëŠ¥í•œ ê´€ì°° ë°ì´í„°ì—ì„œ, **ì„ íƒ í¸í–¥(Selection Bias)**ì„ í†µì œí•˜ê³  ì¸ê³¼ íš¨ê³¼ë¥¼ ì¶”ì •í•˜ëŠ” ëŒ€í‘œì ì¸ ë°©ë²•ì¸ **ê²½í–¥ ì ìˆ˜ ì§ì§“ê¸°(Propensity Score Matching, PSM)**ë¥¼ ì§ì ‘ êµ¬í˜„í•©ë‹ˆë‹¤.


### ğŸ’¡ í•µì‹¬ ê°œë…:

**PSM**ì€ 'íŠ¹ì • ì²˜ì¹˜(treatment)ë¥¼ ë°›ì„ ê²½í–¥'ì„ ì ìˆ˜ë¡œ ê³„ì‚°í•˜ì—¬, ì´ ì ìˆ˜ê°€ ë¹„ìŠ·í•œ ì‹¤í—˜êµ°ê³¼ ëŒ€ì¡°êµ° ê°œì²´ë¥¼ 1:1ë¡œ ì§ì§€ì–´ì£¼ëŠ” ë°©ë²•ì…ë‹ˆë‹¤. ì´ë¥¼ í†µí•´ ì²˜ì¹˜ ì—¬ë¶€ë¥¼ ì œì™¸í•œ ë‹¤ë¥¸ íŠ¹ì„±(ì¤‘ì²©ìš”ì¸)ë“¤ì´ ë¹„ìŠ·í•œ, ì¦‰ **'í†µê³„ì  ìŒë‘¥ì´'** ì§‘ë‹¨ì„ ë§Œë“¤ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì´ 'ì§ì§€ì–´ì§„' ì§‘ë‹¨ ê°„ì˜ ê²°ê³¼ ì°¨ì´ë¥¼ ë¹„êµí•˜ë©´, ì„ íƒ í¸í–¥ì´ ë³´ì •ëœ ìˆœìˆ˜í•œ ì²˜ì¹˜ íš¨ê³¼ì— ë” ê°€ê¹ê²Œ ë‹¤ê°€ê°ˆ ìˆ˜ ìˆìŠµë‹ˆë‹¤.


### ğŸ“Œ ìˆ˜í–‰ ê³¼ì œ:

1.  'ìƒˆë¡œìš´ ì˜¨ë¼ì¸ ê°•ì˜ ìˆ˜ê°• ì—¬ë¶€'ê°€ 'ê¸°ë§ê³ ì‚¬ ì„±ì 'ì— ë¯¸ì¹˜ëŠ” ì˜í–¥ì„ ë¶„ì„í•˜ëŠ” ê°€ìƒ ë°ì´í„°ë¥¼ ìƒì„±í•©ë‹ˆë‹¤. (ë‹¨, 'ì‚¬ì „ ì„±ì 'ì´ ë†’ê³  'í•™ìŠµ ì˜ì§€'ê°€ ë†’ì€ í•™ìƒì¼ìˆ˜ë¡ ê°•ì˜ë¥¼ ì‹ ì²­í•  í™•ë¥ ì´ ë†’ë„ë¡ **ì„ íƒ í¸í–¥**ì„ ë¶€ì—¬í•©ë‹ˆë‹¤.)
2.  **ë‹¨ìˆœ ë¹„êµ**: ì „ì²´ ìˆ˜ê°•ìƒ ì§‘ë‹¨ê³¼ ë¹„ìˆ˜ê°•ìƒ ì§‘ë‹¨ì˜ í‰ê·  ì„±ì ì„ ë¹„êµí•©ë‹ˆë‹¤.
3.  **PSM ìˆ˜í–‰**:
    a. ë¡œì§€ìŠ¤í‹± íšŒê·€ë¥¼ ì´ìš©í•´ ê° í•™ìƒì˜ 'ê°•ì˜ ìˆ˜ê°• ê²½í–¥ ì ìˆ˜(propensity score)'ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
    b. ìˆ˜ê°•ìƒ(ì‹¤í—˜êµ°) í•œ ëª…ë§ˆë‹¤, ê²½í–¥ ì ìˆ˜ê°€ ê°€ì¥ ë¹„ìŠ·í•œ ë¹„ìˆ˜ê°•ìƒ(ëŒ€ì¡°êµ°)ì„ ì§ì§€ì–´ ì¤ë‹ˆë‹¤. (`sklearn.neighbors.NearestNeighbors` í™œìš©)
4.  **ì§ì§€ì–´ì§„ í›„ ë¹„êµ**: 'ì§ì§€ì–´ì§„' ìˆ˜ê°•ìƒ ì§‘ë‹¨ê³¼ ë¹„ìˆ˜ê°•ìƒ ì§‘ë‹¨ì˜ í‰ê·  ì„±ì ì„ ë¹„êµí•˜ê³ , ê·¸ ì°¨ì´(ATT: Average Treatment Effect on the Treated)ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
5.  ë‹¨ìˆœ ë¹„êµ ê²°ê³¼ì™€ PSM í›„ì˜ ê²°ê³¼ë¥¼ ë¹„êµí•˜ê³ , ì™œ ê²°ê³¼ê°€ ë‹¬ë¼ì¡ŒëŠ”ì§€ í•´ì„í•©ë‹ˆë‹¤.


```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.neighbors import NearestNeighbors

# 1. ì„ íƒ í¸í–¥ì´ ìˆëŠ” ë°ì´í„° ìƒì„± (ì´ ì½”ë“œëŠ” ìˆ˜ì •í•˜ì§€ ë§ˆì„¸ìš”)
np.random.seed(42)
n = 1000
pre_score = np.random.normal(60, 10, n)
motivation = np.random.uniform(0, 1, n)
propensity = 1 / (1 + np.exp(-( -5 + (pre_score-60)/5 + motivation*2)))
treatment = np.random.binomial(1, propensity, n)
true_effect = 5
final_score = pre_score + treatment * true_effect + np.random.normal(0, 5, n)
df4 = pd.DataFrame({'pre_score': pre_score, 'motivation': motivation, 'treatment': treatment, 'final_score': final_score})

# 2. ë‹¨ìˆœ ë¹„êµ (Naive Effect)
naive_effect = df4[df4['treatment']==1]['final_score'].mean() - df4[df4['treatment']==0]['final_score'].mean()
print(f"ë‹¨ìˆœ í‰ê·  ë¹„êµ (ì„±ì  ì°¨ì´): {naive_effect:.2f}ì \n")

# 3-a. ê²½í–¥ ì ìˆ˜ ê³„ì‚° (Propensity Score)
X = df4[['pre_score', 'motivation']]
y = df4['treatment']
logit_model = sm.Logit(y, sm.add_constant(X)).fit(disp=0)
df4['propensity_score'] = logit_model.predict(sm.add_constant(X))

# 3-b. ì§ì§“ê¸° (Matching)
treated = df4[df4['treatment'] == 1].copy()
control = df4[df4['treatment'] == 0].copy()

# NearestNeighbors: controlì—ì„œ treatedì™€ propensity_scoreê°€ ê°€ì¥ ê°€ê¹Œìš´ 1ëª… ì°¾ê¸°
nn = NearestNeighbors(n_neighbors=1)  # radiusëŠ” ì—¬ê¸°ì„  í•„ìˆ˜ ì•„ë‹˜
nn.fit(control[['propensity_score']])

distances, indices = nn.kneighbors(treated[['propensity_score']])
matched_control = control.iloc[indices.flatten()].copy()

# 4. ì§ì§€ì–´ì§„ í›„ ë¹„êµ (ATT ê³„ì‚°)
att_effect = treated['final_score'].mean() - matched_control['final_score'].mean()
print(f"PSM ì´í›„ í‰ê·  ë¹„êµ (ì¶”ì •ëœ ì¸ê³¼ íš¨ê³¼, ATT): {att_effect:.2f}ì \n")

# (ì„ íƒ) ë§¤ì¹­ í’ˆì§ˆ ê°„ë‹¨ ì²´í¬: ë§¤ì¹­ í›„ propensity_score ì°¨ì´ í‰ê· 
print(f"ë§¤ì¹­ í›„ í‰ê·  propensity_score ì°¨ì´: {(treated['propensity_score'].values - matched_control['propensity_score'].values).mean():.4f}")

# 5. ê²°ê³¼ í•´ì„ (ì•„ë˜ ì£¼ì„ì— ì§ì ‘ ì‘ì„±í•´ ë³´ì„¸ìš”)
# ë‹¨ìˆœ ë¹„êµ ê²°ê³¼ê°€ ì‹¤ì œ íš¨ê³¼(5ì )ë³´ë‹¤ í¬ê²Œ ë‚˜ì˜¨ ì´ìœ ëŠ”?:
# - ìˆ˜ê°•ìƒì€ ì›ë˜(pre_scoreê°€ ë†’ê³  motivationì´ ë†’ì€) í•™ìƒì¼ í™•ë¥ ì´ ë” ì»¤ì„œ,
#   ê°•ì˜ íš¨ê³¼(5ì ) + ì›ë˜ ì˜í•˜ëŠ”/ì—´ì‹¬íˆ í•˜ëŠ” íŠ¹ì„±(ì„ íƒ í¸í–¥)ê¹Œì§€ í•¨ê»˜ ë°˜ì˜ë˜ì–´ ì„±ì ì´ ë” ë†’ê²Œ ê´€ì¸¡ë˜ì—ˆê¸° ë•Œë¬¸ì…ë‹ˆë‹¤.
#
# PSMì„ í†µí•´ ì´ ë¬¸ì œë¥¼ ì–´ë–»ê²Œ í•´ê²°í•  ìˆ˜ ìˆì—ˆë‚˜ìš”?:
# - pre_scoreì™€ motivationì„ ì´ìš©í•´ ê° í•™ìƒì˜ ìˆ˜ê°• í™•ë¥ (ê²½í–¥ì ìˆ˜)ì„ ê³„ì‚°í•œ ë’¤,
#   ê²½í–¥ì ìˆ˜ê°€ ë¹„ìŠ·í•œ ë¹„ìˆ˜ê°•ìƒì„ ìˆ˜ê°•ìƒê³¼ 1:1ë¡œ ì§ì§€ì–´ ë¹„êµí•¨ìœ¼ë¡œì¨
#   ë‘ ì§‘ë‹¨ì˜ ì‚¬ì „ ì¡°ê±´(pre_score, motivation)ì„ ìµœëŒ€í•œ ë¹„ìŠ·í•˜ê²Œ ë§ì·„ìŠµë‹ˆë‹¤.
#   ê·¸ ê²°ê³¼, ë‹¨ìˆœ ë¹„êµì—ì„œ ì„ì—¬ ìˆë˜ â€˜ì„ íƒ í¸í–¥â€™ì„ ì¤„ì´ê³  ê°•ì˜ì˜ ìˆœìˆ˜í•œ íš¨ê³¼(ATT)ë¥¼ ë” ê°€ê¹ê²Œ ì¶”ì •í•  ìˆ˜ ìˆì—ˆìŠµë‹ˆë‹¤.
```

### ğŸ¤” ìƒê°í•´ ë³¼ ë¬¸ì œ:

ì§ì§“ê¸°(Matching)ê°€ ì˜ ë˜ì—ˆëŠ”ì§€ í™•ì¸í•˜ëŠ” ë°©ë²• ì¤‘ í•˜ë‚˜ëŠ”, ì§ì§€ì–´ì§„ ë‘ ì§‘ë‹¨ì˜ ì¤‘ì²©ìš”ì¸(ì—¬ê¸°ì„œëŠ” `pre_score`, `motivation`) í‰ê· ì´ ì„œë¡œ ë¹„ìŠ·í•´ì¡ŒëŠ”ì§€ í™•ì¸í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤. ì´ë¥¼ **'ê³µë³€ëŸ‰ ê· í˜•(Covariate Balance)'** í™•ì¸ì´ë¼ê³  í•©ë‹ˆë‹¤. ì§ì§“ê¸° ì „ê³¼ í›„ì˜ `pre_score` í‰ê·  ì°¨ì´ë¥¼ ê°ê° ê³„ì‚°í•´ë³´ê³ , PSMì´ ê³µë³€ëŸ‰ ê· í˜•ì„ ë§ì¶”ëŠ” ë° ì„±ê³µí–ˆëŠ”ì§€ í™•ì¸í•´ë³´ì„¸ìš”.

### ğŸ’¡ í•µì‹¬ ê°œë…:

**ì´ì¤‘ì°¨ë¶„ë²•(DiD)**ì€ ì •ì±…ì´ ì‹œí–‰ëœ **ì‹¤í—˜êµ°(Treatment Group)**ê³¼ ì‹œí–‰ë˜ì§€ ì•Šì€ **í†µì œêµ°(Control Group)**ì˜ **ì •ì±… ì‹œí–‰ ì „í›„(Pre-Post)** ë°ì´í„°ë¥¼ ëª¨ë‘ í™œìš©í•˜ëŠ” ê°•ë ¥í•œ ì¸ê³¼ì¶”ë¡  ë°©ë²•ì…ë‹ˆë‹¤. í•µì‹¬ ì•„ì´ë””ì–´ëŠ” 'ì •ì±…ì´ ì—†ì—ˆì„ ê²½ìš°ì˜ ìì—°ì ì¸ ì‹œê°„ ë³€í™”'ë¥¼ í†µì œêµ°ì„ í†µí•´ ì¶”ì •í•˜ê³ , ì´ë¥¼ ì‹¤í—˜êµ°ì˜ ë³€í™”ì—ì„œ ë¹¼ ì¤Œìœ¼ë¡œì¨ ì •ì±…ì˜ ìˆœìˆ˜í•œ íš¨ê³¼ë§Œ ë°œë¼ë‚´ëŠ” ê²ƒì…ë‹ˆë‹¤. íšŒê·€ë¶„ì„ì—ì„œëŠ” `(ì‹¤í—˜êµ° ì—¬ë¶€) * (ì •ì±… í›„ ì—¬ë¶€)`ì˜ **ìƒí˜¸ì‘ìš© í•­** ê³„ìˆ˜ê°€ ë°”ë¡œ DiD ì¶”ì •ì¹˜ê°€ ë©ë‹ˆë‹¤.


### ğŸ“Œ ìˆ˜í–‰ ê³¼ì œ:

1.  'ì„œìš¸ì‹œ'ì—ë§Œ 'ë¬´ë£Œ ê³µê³µ ì™€ì´íŒŒì´' ì •ì±…ì„ ë„ì…í–ˆì„ ë•Œ, ì‹œë¯¼ë“¤ì˜ 'ì›” í‰ê·  ë°ì´í„° ì‚¬ìš©ëŸ‰'ì— ë¯¸ì¹˜ëŠ” ì˜í–¥ì„ ë¶„ì„í•˜ëŠ” ê°€ìƒ ë°ì´í„°ë¥¼ ìƒì„±í•©ë‹ˆë‹¤. (ì‹¤í—˜êµ°: ì„œìš¸, í†µì œêµ°: ë¶€ì‚°)
2.  **ì‹œê°í™”**: ì •ì±… ì‹œí–‰ ì „í›„, ë‘ ë„ì‹œì˜ í‰ê·  ë°ì´í„° ì‚¬ìš©ëŸ‰ ë³€í™”ë¥¼ êº¾ì€ì„  ê·¸ë˜í”„ë¡œ ê·¸ë ¤ë´…ë‹ˆë‹¤.
3.  **DiD íšŒê·€ë¶„ì„**: `ë°ì´í„°ì‚¬ìš©ëŸ‰ ~ is_ì„œìš¸ + is_ì •ì±…í›„ + is_ì„œìš¸*is_ì •ì±…í›„` í˜•íƒœì˜ íšŒê·€ ëª¨ë¸ì„ ë§Œë“­ë‹ˆë‹¤.
    - `is_ì„œìš¸`: ì„œìš¸ì´ë©´ 1, ë¶€ì‚°ì´ë©´ 0
    - `is_ì •ì±…í›„`: ì •ì±… ì‹œí–‰ í›„(2024ë…„)ì´ë©´ 1, ì‹œí–‰ ì „(2023ë…„)ì´ë©´ 0
    - `is_ì„œìš¸*is_ì •ì±…í›„`: ìƒí˜¸ì‘ìš© í•­
4.  íšŒê·€ë¶„ì„ ê²°ê³¼í‘œì—ì„œ ìƒí˜¸ì‘ìš© í•­ì˜ ê³„ìˆ˜(coef)ì™€ p-valueë¥¼ í™•ì¸í•˜ê³ , ì •ì±…ì˜ íš¨ê³¼ë¥¼ í•´ì„í•´ ë³´ì„¸ìš”.


```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

# 1. DiD ë¶„ì„ì„ ìœ„í•œ ê°€ìƒ íŒ¨ë„ ë°ì´í„° ìƒì„± (ìˆ˜ì • ê¸ˆì§€)
np.random.seed(2025)
n_per_group = 500
df_seoul_pre = pd.DataFrame({'city': 'ì„œìš¸', 'year': 2023, 'data_usage': np.random.normal(15, 3, n_per_group)})
df_seoul_post = pd.DataFrame({'city': 'ì„œìš¸', 'year': 2024, 'data_usage': df_seoul_pre['data_usage'] + 2 + 5 + np.random.normal(0, 1, n_per_group)})
df_busan_pre = pd.DataFrame({'city': 'ë¶€ì‚°', 'year': 2023, 'data_usage': np.random.normal(10, 3, n_per_group)})
df_busan_post = pd.DataFrame({'city': 'ë¶€ì‚°', 'year': 2024, 'data_usage': df_busan_pre['data_usage'] + 2 + np.random.normal(0, 1, n_per_group)})
df5 = pd.concat([df_seoul_pre, df_seoul_post, df_busan_pre, df_busan_post])

# 2. ì‹œê°í™” (ìˆ˜ì • ê¸ˆì§€)
sns.lineplot(data=df5, x='year', y='data_usage', hue='city', marker='o', errorbar='ci')
plt.title('Average Data Usage by City Before and After Policy Implementation')
plt.xticks([2023, 2024]); plt.grid(True, axis='y', linestyle='--')
plt.show()

# 3. DiD íšŒê·€ë¶„ì„ì„ ìœ„í•œ ë³€ìˆ˜ ìƒì„±
df5['is_seoul'] = (df5['city'] == 'ì„œìš¸').astype(int)
df5['is_post'] = (df5['year'] == 2024).astype(int)
df5['did_term'] = df5['is_seoul'] * df5['is_post']

y = df5['data_usage']
X = df5[['is_seoul', 'is_post', 'did_term']]

# OLS íšŒê·€ëª¨ë¸ í•™ìŠµ
model_did = sm.OLS(y, sm.add_constant(X)).fit()
print(model_did.summary())
```

ì´ì¤‘ì°¨ë¶„ë²•ì´ ì„±ë¦½í•˜ê¸° ìœ„í•œ ê°€ì¥ ì¤‘ìš”í•œ ê°€ì •ì€ **'í‰í–‰ì¶”ì„¸ê°€ì •(Parallel Trends Assumption)'**ì…ë‹ˆë‹¤. ì¦‰, 'ë§Œì•½ ì •ì±…ì´ ì—†ì—ˆë”ë¼ë©´, ì„œìš¸ê³¼ ë¶€ì‚°ì˜ ë°ì´í„° ì‚¬ìš©ëŸ‰ì€ ê³„ì† í‰í–‰í•˜ê²Œ ì¦ê°€í–ˆì„ ê²ƒì´ë‹¤'ë¼ëŠ” ê°€ì •ì…ë‹ˆë‹¤. ìœ„ ì‹œê°í™” ê·¸ë˜í”„ì—ì„œ ì •ì±… ì‹œí–‰ **ì „** ì‹œì (2023ë…„ ì´ì „)ì˜ ë°ì´í„°ê°€ ì—¬ëŸ¬ ê°œ ë” ìˆë‹¤ë©´, ì´ ê°€ì •ì„ ì–´ë–»ê²Œ í™•ì¸í•  ìˆ˜ ìˆì„ì§€ ì„¤ëª…í•´ ë³´ì„¸ìš”.

ì •ì±… ì‹œí–‰ ì „ ì‹œì ì˜ ë°ì´í„°ê°€ ì—¬ëŸ¬ ê°œ ì¡´ì¬í•œë‹¤ë©´, ì„œìš¸ê³¼ ë¶€ì‚°ì˜ í‰ê·  ë°ì´í„° ì‚¬ìš©ëŸ‰ ì¶”ì´ê°€ ì •ì±… ì´ì „ ê¸°ê°„ ë™ì•ˆ ì„œë¡œ ê±°ì˜ í‰í–‰í•œ í˜•íƒœë¡œ ì›€ì§ì˜€ëŠ”ì§€ë¥¼ í™•ì¸í•¨ìœ¼ë¡œì¨ í‰í–‰ì¶”ì„¸ê°€ì •ì„ ì ê²€í•  ìˆ˜ ìˆë‹¤. êµ¬ì²´ì ìœ¼ë¡œëŠ” ì •ì±… ì´ì „ ì—¬ëŸ¬ ì—°ë„ì— ëŒ€í•´ ë‘ ë„ì‹œì˜ í‰ê·  ë°ì´í„° ì‚¬ìš©ëŸ‰ì„ ì‹œê°í™”í–ˆì„ ë•Œ ì¦ê°€ ë°©í–¥ê³¼ ê¸°ìš¸ê¸°(ë³€í™”ìœ¨)ê°€ ìœ ì‚¬í•˜ê³ , ë‘ ì„  ê°„ì˜ ê°„ê²©ì´ ì¼ì •í•˜ê²Œ ìœ ì§€ë˜ëŠ”ì§€ë¥¼ í™•ì¸í•˜ë©°, ì •ì±… ì‹œí–‰ ì´ì „ì— ë‘ ë„ì‹œ ê°„ ê²©ì°¨ê°€ ì²´ê³„ì ìœ¼ë¡œ ë²Œì–´ì§€ê±°ë‚˜ ì¢í˜€ì§€ëŠ” ì¶”ì„¸ê°€ ì—†ë‹¤ë©´ í‰í–‰ì¶”ì„¸ê°€ì •ì´ íƒ€ë‹¹í•˜ë‹¤ê³  íŒë‹¨ë¨
